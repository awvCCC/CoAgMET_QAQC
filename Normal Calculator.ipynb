```python
# Loads the libraries required to build the environment we will be using in this tutorial
import os
import urllib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
from scipy import stats
import calendar
```

```python
path = r'C:\Users\student\Documents\Alistair CoAgMET Projects\Pythonstuff\QC\minimums\CoAgMET_minimums'
os.chdir(path)
xl_file = r"C:\Users\student\Documents\Alistair CoAgMET Projects\Pythonstuff\QC\colocation_metadata.xlsx"
csv = r'C:\Users\student\Documents\Alistair CoAgMET Projects\Pythonstuff\QC\normals\COOP_Normals\csv.csv'

xl = pd.ExcelFile(xl_file)

meta = xl.parse('metadata')
meta['sdate'] = meta.sdate.dt.date
meta['edate'] = meta.edate.dt.date
meta.drop(meta.columns[3],axis=1)

n = 0
for index, rows in meta.iterrows():
    print('Grabbing data for ' + str(meta.ID[n]))
    urllib.request.urlretrieve(
        'http://coagmet.colostate.edu/cgi-bin/web_services.pl?type=daily&sids=' + str(meta.ID[n]) + '&sdate=' + str(meta.sdate[n]) + '&edate=' + str(meta.edate[n]) + '&elems=tave,tmax,tmin,vp,rhmax,rhmin,sr,wrun,pp,st5mx,st5mn,st15mx,st15mn',
        filename=csv)
    df = pd.read_csv(csv)

    df = df.reset_index()
    df.columns.values[0] = 'station'
    df.columns.values[1] = 'date'
    df.columns.values[2] = 'tmean'
    df.columns.values[3] = 'tmax'
    df.columns.values[4] = 'tmin'
    df.columns.values[5] = 'vp'
    df.columns.values[6] = 'rhmax'
    df.columns.values[7] = 'rhmin'
    df.columns.values[8] = 'sr'
    df.columns.values[9] = 'wrun'
    df.columns.values[10] = 'pp'
    df.columns.values[11] = 'st5mx'
    df.columns.values[12] = 'st5mn'
    df.columns.values[13] = 'st15mx'
    df.columns.values[14] = 'st15mn'

    df['date'] = pd.to_datetime(df.date)
    df['Julian'] = df.date.dt.strftime('%j')
    normals = df.groupby(['Julian']).min().reset_index()
    normals['pp'] = normals['pp'].cumsum()
    normals.pp = normals.pp / 25.4
    normals.tmean = (9.0/5.0) * (normals.tmean + 32)
    normals.tmax = (9.0/5.0) * (normals.tmax + 32)
    normals.tmin = (9.0/5.0) * (normals.tmin + 32)
    normals.st5mx = (9.0/5.0) * (normals.st5mx + 32)
    normals.st5mn = (9.0/5.0) * (normals.st5mn + 32)
    normals.st15mx = (9.0/5.0) * (normals.st15mx + 32)
    normals.st15mn = (9.0/5.0) * (normals.st15mn + 32)
    normals.wrun = normals.wrun / 1.609344
    normals.to_csv(path + r'/' + str(meta.ID[n]) + 'min.csv')
    n = n + 1
```
