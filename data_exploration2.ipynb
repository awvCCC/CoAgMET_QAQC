```python
import os
import urllib
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
```

```python
path = r'C:\Users\student\Documents\Alistair CoAgMET Projects\Pythonstuff\QC'
os.chdir(path)

csv = r'C:\Users\student\Documents\Alistair CoAgMET Projects\Pythonstuff\QC\csv.csv'
f = open(csv,'w+')

txt = (r'C:\Users\student\Documents\Alistair CoAgMET Projects\Pythonstuff\QC\data_exploration\daily' + 
       r'\akr02.txt')
g = open(txt,'w+')

empty_line = ('\n')
```

```python
empty_line = str('\n')
print('Pulling data...')
urllib.request.urlretrieve(
    'http://coagmet.colostate.edu/cgi-bin/web_services.pl?type=daily&sids=akr02&sdate=1992-07-01' + 
    '&edate=2018-12-31' + 
    '&elems=tave,tmax,tmin,vp,rhmax,rhmin,sr,wrun,pp,st5mx,st5mn,st15mx,st15mn,volts,gust,gustdir',
    filename=csv)
print('Data pull was successful.')
print(empty_line)

print('Loading csv into pandas dataframe...')
df = pd.read_csv(csv)
df = df.reset_index()
print('CSV successfully loaded into a dataframe.')
print(empty_line)

print('Preparing dataframe for analysis...')
df.columns = ['station','date','tave','tmax','tmin','vp','rhmax','rhmin','sr','wrun','pp','st5mx',
             'st5mn','st15mx','st15mn','volts','gust','gustdir']
df['date'] = pd.to_datetime(df.date)
df.drop(df.tail(1).index, inplace=True)
cols = df.columns.drop(['station','date'])
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
print('Dataframe is ready for analysis.')
print(empty_line)

print('Writing a statistical summary of the raw data to a text file...')
title = str('Summary of Data Exploration of AKR02 for Period of Record')
g.write(title.center(24))
g.write(empty_line)
g.write(empty_line)
g.write(str('Statistical summary of Raw Data:'))
g.write(empty_line)
g.write(empty_line)
raw_summary = df.describe()
g.write(str(raw_summary) + empty_line)
g.write(empty_line)
print('Text file successfully appended')
print(empty_line)

og_length = len(df.index)
print('Raw dataset containes' + str(og_length) + ' observations.')
print('')

print('Statistical Summary of Raw Data:')
print(raw_summary)
print(empty_line)

null_obs_string = ('Summary of null observations found in the raw dataset:')
g.write(null_obs_string)
g.write(empty_line)

print('Checking for null values reported in the average daily temperature column...')
d = 0
tave_nulls = 0
for index,rows in df.iterrows():
    if(pd.isnull(df['tave'][d])):
        tave_nulls = tave_nulls + 1
    d = d + 1
tave_string = ('There are ' + str(tave_nulls) + 
      ' null observations of daily average temperature in the dataset.')
g.write(tave_string)
g.write(empty_line)
print(tave_string)
print(empty_line)

print('Checking for null values reported in the daily maximum temperature column...')
print(empty_line)
d = 0 
tmax_nulls = 0
for index,rows in df.iterrows():
    if(pd.isnull(df['tmax'][d])):
        tmax_nulls = tmax_nulls + 1
    d = d + 1
tmax_string = ('There are ' + str(tmax_nulls) + 
      ' null observations of daily maximum temperature in the dataset.')
g.write(tmax_string)
g.write(empty_line)
print(tmax_string)
print(empty_line)

print('Checking for null values reported in the daily minimum temperature column...')
print('')
d = 0
tmin_nulls = 0
for index,rows in df.iterrows():
    if (pd.isnull(df['tmin'][d])):
        tmin_nulls = tmin_nulls + 1
    d = d + 1
tmin_string = ('There are ' + str(tmin_nulls) + 
      ' null observations of daily minimum temperature in the dataset.')
g.write(tmin_string)
g.write(empty_line)
print(tmin_string)
print(empty_line)

print('Checking for null values reported in the daily average vapor pressure column...')
d = 0
vp_nulls = 0
for index,rows in df.iterrows():
    if (pd.isnull(df['vp'][d])):
        vp_nulls = vp_nulls + 1
    d = d + 1
vp_string = ('There are ' + str(vp_nulls) + 
     ' null observations of daily average vapor pressure in the dataset')
g.write(vp_string)
g.write(empty_line)
print(vp_string)
print(empty_line)

print('Checking for null values reported in the daily maximum relative humidity column...')
d = 0 
rhmax_nulls = 0
for index,rows in df.iterrows():
    if(pd.isnull(df['rhmax'][d])):
        rhmax_nulls = rhmax_nulls + 1
    d = d + 1
rhmax_string = ('There are ' + str(rhmax_nulls) + 
     ' null observations of daily maximum relative humidity in the dataset.')
g.write(rhmax_string)
g.write(empty_line)
print(rhmax_string)
print(empty_line)

print('Checking for null values reported in the daily minimum relative humidity column...')
print('')
d = 0
rhmin_nulls = 0
for index,rows in df.iterrows():
    if(pd.isnull(df['rhmin'][d])):
        rhmin_nulls = rhmin_nulls + 1
    d = d + 1
rhmin_string = ('There are ' + str(rhmin_nulls) + 
      ' null observations of daily minimum relative humidity in the dataset.')
g.write(rhmin_string)
g.write(empty_line)
print(rhmin_string)
print(empty_line)

print('Checking for null values reported in the daily average solar radiation column...')
print('')
d = 0
sr_nulls = 0
for index,rows in df.iterrows():
    if(pd.isnull(df['sr'][d])):
        sr_nulls = sr_nulls + 1
    d = d + 1
sr_string = ('There are ' + str(sr_nulls) + 
      ' null observations of daily average solar radiation in the dataset.')
g.write(sr_string)
g.write(empty_line)
print(sr_string)
print(empty_line)

print('Checking for null values reported in the daily total windrun column...')
print('')
d = 0 
wrun_nulls = 0
for index,rows in df.iterrows():
    if(pd.isnull(df['wrun'][d])):
        if(pd.isnull(df['wrun'][d])):
            wrun_nulls = wrun_nulls + 1
        d = d + 1
wrun_string = ('There are ' + str(wrun_nulls) + 
     ' null observations of daily total wind run in the dataset.')
g.write(wrun_string)
g.write(empty_line)
print(wrun_string)
print(empty_line)

print('Checking for null values reported in the daily total precipitation column...')
print('')
d = 0
pp_nulls = 0
for index,rows in df.iterrows():
    if(pd.isnull(df['pp'][d])):
        pp_nulls = pp_nulls + 1
    d = d + 1
pp_string = str('There are ' + str(pp_nulls) + 
     ' null observations of daily total precipitation in the dataset.')
g.write(pp_string)
g.write(empty_line)
print(pp_string)
print(empty_line)

print('Checking for null values reported in the daily maximum 5cm soil temperature column...')
print('')
d = 0
st5mx_nulls = 0
for index,rows in df.iterrows():
    if(pd.isnull(df['st5mx'][d])):
        st5mx_nulls = st5mx_nulls + 1
    d = d + 1
st5mx_string = ('There are ' + str(st5mx_nulls) + 
     ' null observations of daily maximum 5cm soil temperature in the dataset.')
g.write(st5mx_string)
g.write(empty_line)
print(st5mx_string)
print(empty_line)

print('Checking for null values reported in the daily minimum 5cm soil temperature column...')
d = 0
st5mn_nulls = 0
for index,rows in df.iterrows():
    if(pd.isnull(df['st5mn'][d])):
        st5mn_nulls = st5mn_nulls + 1
    d = d + 1
st5mn_string = ('There are ' + str(st5mn_nulls) +
     ' null observations of daily minimum 5cm soil temperature in the dataset.')
g.write(st5mn_string)
g.write(empty_line)
print(st5mn_string)
print(empty_line)

print('Checking for null values reported in the daily maximum 15cm soil temperature column...')
d = 0
st15mx_nulls = 0
for index,rows in df.iterrows():
    if(pd.isnull(df['st15mx'][d])):
        st15mx_nulls = st15mx_nulls + 1
    d = d + 1
st15mx_string = ('There are ' + str(st15mx_nulls) + 
                 ' null observations of daily maximum 15cm soil temperature in the dataset.')
g.write(st15mx_string)
g.write(empty_line)
print(st15mx_string)
print(empty_line)

print('Checking for null values reported in the daily minimum 15cm soil temperature column...')
d = 0
st15mn_nulls = 0 
for index,rows in df.iterrows():
    if(pd.isnull(df['st15mn'][d])):
        st15mn_nulls = st15mn_nulls + 1
    d = d + 1
st15mn_string = ('There are ' + str(st15mn_nulls) + 
                 ' null observations of daily minimum 15cm soil temperature in the dataset.')
g.write(st15mn_string)
g.write(empty_line)
print(st15mn_string)
print(empty_line)

print('Checking for null values reported in the daily average battery voltage column...')
d = 0
volts_nulls = 0
for index,rows in df.iterrows():
    if(pd.isnull(df['volts'][d])):
        volts_nulls = volts_nulls + 1
    d = d + 1
volts_string = ('There are ' + str(volts_nulls) + 
     ' null observations of daily average batter voltage in the dataset.')
g.write(volts_string)
g.write(empty_line)
print(volts_string)
print(empty_line)

print('Checking for null values reported in daily maximum gust speed column...')
d = 0
gust_nulls = 0
for index,rows in df.iterrows():
    if(pd.isnull(df['gust'][d])):
        gust_nulls = gust_nulls + 1
    d = d + 1
gust_string = ('There are ' + str(gust_nulls) + 
     ' null observations of daily maximum gust speed in the dataset.')
g.write(gust_string)
g.write(empty_line)
print(gust_string)
print(empty_line)

print('Checking for null values reported in the daily maximum gust direction column...')
d = 0
gustdir_nulls = 0
for index,rows in df.iterrows():
    if(pd.isnull(df['gustdir'][d])):
        gustdir_nulls = gustdir_nulls + 1
    d = d + 1
gustdir_string = ('There are ' + str(gustdir_nulls) + 
     ' null observations of daily maximum gust direction in the dataset.')
g.write(gustdir_string)
g.write(empty_line)
print(gustdir_string)
print(empty_line)

print('Repopulating the dataset to find any observations that are entirely missing...')
r = pd.date_range(start = '1992-07-01',end='2018-12-31',freq='D')
missing_obs = len(r) - og_length
missing_obs_string = ('There are ' + str(missing_obs) + 
                      ' observations entirely missing fomr the dataset.')
g.write(missing_obs_string)
g.write(empty_line)
print(missing_obs_string)
df = df.set_index('date').reindex(r,copy=False).rename_axis('date')
print('Repopulation successul. All observations that were previously missing from ' + 
     'the dataset are now considered null observations')
print(empty_line)

print('Salvaging missing daily average temperature data were possible...')
d = 0
for index,rows in df.iterrows():
    while (pd.isnull(df['tave'][d])) and (pd.notnull(df['tmax'][d])) and (pd.notnull(df['tmin'][d])):
        df.tave = (df.tmin + df.tmax) / 2
    d = d + 1
print('Salvage was successful.')
print(empty_line)

clean_summary = df.describe()
print('Statistical Summary of Clean Data:')
print(clean_summary)
g.write('Statistical Summary of Clean Data:')
g.write(empty_line)
g.write(str(clean_summary))
g.close()

nulls_dict = ({'tave':[tave_nulls],'tmax':[tmax_nulls],'tmin':[tmin_nulls],'vp':[vp_nulls],
              'rhmax':[rhmax_nulls],'rhmin':[rhmin_nulls],'sr':[sr_nulls],'wrun':[wrun_nulls],
              'pp':[pp_nulls],'st5mx':[st5mx_nulls],'st5mn': [st5mn_nulls],'st15mx':[st15mx_nulls],
              'st15mn':[st15mn_nulls],'volts':[volts_nulls],'gust':[gust_nulls],
              'gustdir':[gustdir_nulls]})

nulls_df = pd.DataFrame(nulls_dict)
nulls_df

        
```
