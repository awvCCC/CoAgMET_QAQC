```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import urllib
from jupyterthemes import jtplot
```

```python
path = (r'C:\Users\student\Documents\Alistair CoAgMET Projects\Pythonstuff\PrecipComparison' + 
        r'\variable_isolation\hourly')
os.chdir(path)

coag_csv = (path + r'\coag_csv.csv')
coop_csv = (path + r'\coop_csv.csv')
g = open(coop_csv,'w+')
f = open(coag_csv,'w+')
jtplot.style(theme='gruvboxd')
```

```python
urllib.request.urlretrieve(
    'http://coagmet.colostate.edu/cgi-bin/web_services.pl?type=hourly&sids=rfd01&sdate=1992-06-04' + 
    '&edate=2018-12-31&elems=tmean,rh,vp,sr,ws,wind_vec,pp,gust,gustdir', filename=coag_csv)
coag_df = pd.read_csv(coag_csv)
coag_df = coag_df.reset_index()
```

```python
coag_df.columns = ['station','datetime','tmean','rh','vp','sr','ws','wind_vec','pp','gust','gustdir']
coag_df.drop(coag_df.tail(1).index, inplace=True)
coag_df = coag_df.drop(coag_df.columns[0],axis=1)
coag_df['datetime'] = pd.to_datetime(coag_df.datetime, errors='coerce')
#coag_df['j_day'] = coag_df.datetime.dt.strftime('%j').astype(int)
```

```python
pp_daily_sum = coag_df.resample(rule='24H', label='right',on='datetime', base=17).agg(pd.Series.sum,
                                                                                 skipna=False)
pp_cols = ['tmean','rh','vp','sr','ws','wind_vec','gust','gustdir']
pp_daily_sum = pp_daily_sum.drop(pp_cols,axis=1)
pp_daily_sum['j_day'] = pp_daily_sum.index.strftime('%j').astype(int)
pp_daily_sum
```

```python
first_day = pp_daily_sum['j_day'][0]
last_day = pp_daily_sum['j_day'][-1]
```

```python
is_raining = coag_df['pp'] > 0
coag_df_raining = coag_df[is_raining]
coag_df_raining
```

```python
coag_df_raining['datetime'] = pd.to_datetime(coag_df_raining.datetime)
mean_daily_variables = coag_df_raining.resample(rule='24H',label='right',on='datetime',base=17).agg(pd.Series.mean, skipna=False)
mean_daily_variables
```

```python
mean_daily_variables['j_day'] = mean_daily_variables.index.strftime('%j').astype(int)
```

```python
stuff = ['datetime','pp']
mean_daily_variables = mean_daily_variables.drop(stuff,axis=1)
mean_daily_variables
```

```python
first_rainy_day = mean_daily_variables.j_day[0]
last_rainy_day = mean_daily_variables.j_day[-1]
```

```python
start = int(first_rainy_day - first_day)
end = int(last_day - last_rainy_day)
print(start)
print(end)
```

```python
pp_daily_sum.drop(pp_daily_sum.head(start).index,inplace=True)
pp_daily_sum.drop(pp_daily_sum.tail(end).index,inplace=True)
pp_daily_sum
```

```python
new_df = pd.merge(mean_daily_variables,pp_daily_sum,on=pp_daily_sum.index)
new_df
```

```python
#is_raining = new_df['pp'] > 0
#new_df = new_df[is_raining]
new_df.pp = new_df.pp / 25.4
new_df['key_0'] = pd.to_datetime(new_df.key_0, errors='coerce')
new_df['key_0'] = new_df.key_0.dt.date
stuff = ['j_day_x','j_day_y']
new_df.drop(stuff,axis=1,inplace=True)
new_df
```

```python
urllib.request.urlretrieve(
    'http://data.rcc-acis.org/StnData?sid=057167&sdate=1992-06-04&edate=2018-12-31&elems=pcpn&output=csv',
    filename=coop_csv)
coop_df = pd.read_csv(coop_csv)
coop_df = coop_df.reset_index()
coop_df = coop_df.replace('M', np.nan, regex=True)
coop_df = coop_df.replace('T',0.005, regex=True)
coop_df.columns = ['date','precip']
coop_df['date'] = pd.to_datetime(coop_df.date)
coop_df.drop(coop_df.head(start).index,inplace=True)
coop_df.drop(coop_df.tail(end).index,inplace=True)
```

```python
newest_df = new_df.merge(coop_df,left_index=True,right_index=True)
stuff = ['date']
newest_df = newest_df.drop(stuff,axis=1)
newest_df.columns = ['date','tmean','rh','vp','sr','ws','wind_vec',
                     'gust','gustdir','coag_pp','coop_pp']
```

```python
newest_df.coag_pp = newest_df.coag_pp.astype(float)
newest_df.coop_pp = newest_df.coop_pp.astype(float)
dat_newnew = newest_df[(newest_df.coag_pp > 0)]
dat_newnew['diff'] = dat_newnew.coag_pp - dat_newnew.coop_pp
dat_newnew
```
