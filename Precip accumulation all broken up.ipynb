```python
import os
from datetime import datetime,timedelta
import urllib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
today = datetime.now().strftime("%Y-%m-%d")
path = r'C:\Users\student\Documents\Alistair CoAgMET Projects\Pythonstuff\QC\precip_QC\2019WY'

os.chdir(path)
os.mkdir(path + r'\accumulated_precip_' + str(today))

figpath = r'C:\Users\student\Documents\Alistair CoAgMET Projects\Pythonstuff\QC\precip_QC\2019WY\accumulated_precip_' + str(today)

csv = r'C:\Users\student\Documents\Alistair CoAgMET Projects\Pythonstuff\QC\precip_QC\csv.csv'
f = open(csv,'w+')
wy_sdate = '2018-10-01'

sns.set_style('dark')
sns.set(rc={'figure.figuresize' : (11.7,8.27)})
```

```python
urllib.request.urlretrieve('http://coagmet.colostate.edu/cgi-bin/web_services.pl?type=daily&sids=' + str(arkCollegiateValley)
                          + '&sdate=' + str(wy_sdate) + '&edate=' + str(today) + '&elems=pp',
                         filename=csv)
df = pd.read_csv(csv)
f.close()
df = df.reset_index()
df.columns.values[0] = 'station'
df.columns.values[1] = 'date'
df.columns.values[2] = 'precip'
df.precip.astype(float)
df.precip = df.precip / 25.4

df['accumulated_precip'] = df.groupby('station')['precip'].cumsum()

plot = sns.lineplot(x=df.date,y=df.accumulated_precip,hue=df.station,data=df,palette='colorblind')
plot.set(xlabel='Date',ylabel='Accumulated Precip for the 2019 WY')
title = str('Arkansas Drainage - Collegiate Valley')
plot.xaxis.set_major_locator(plt.MaxNLocator(6))
plt.savefig(str(figpath) + r'\Arkansas_Collegiate_Valley_' + str(today) + '.png')
plt.close()
```
