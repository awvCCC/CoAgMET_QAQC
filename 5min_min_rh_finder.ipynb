```python
import os
import pandas as pd
import urllib
from datetime import datetime,timedelta
```

```python
path = r'C:\Users\student\Documents\Alistair CoAgMET Projects\Pythonstuff\QC\relative_humidity'
csv = r'C:\Users\student\Documents\Alistair CoAgMET Projects\Pythonstuff\QC\relative_humidity\csv.csv'

os.chdir(path)
f = open(csv,'w+')

today = datetime.now().strftime("%Y-%m-%d")
yesterday = datetime.strftime(datetime.now() - timedelta(days=1), "%Y-%m-%d")
yesterday
```

```python
urllib.request.urlretrieve(
    'http://coagmet.colostate.edu/cgi-bin/web_services.pl?type=five_minute&sids=ftc01&sdate=' + str(yesterday) + '&edate=' + str(today) + '&elems=rh',
    filename=csv)
df = pd.read_csv(csv)
f.close()
df=df.reset_index()
df.columns.values[0] = 'station'
df.columns.values[1] = 'date'
df.columns.values[2] = 'rh'
df
```

```python
rhmin = df['rh'].min()
print('Yesterdays minimum relative humidity at ftc01 was ' + str(rhmin))
```
