```python
urllib.request.urlretrieve(
    'http://coagmet.colostate.edu/cgi-bin/web_services.pl?type=daily&sids=akr02&sdate=1992-07-01&edate=2018-12-31&elems=pp',
    filename=csv)
df = pd.read_csv(csv)
df = df.reset_index()
df.columns = ['station','date','pp']
df.drop(df.tail(1).index,inplace=True)
df['date'] = pd.to_datetime(df.date)
df['julian'] = df.date.dt.strftime('%j')
df['new_julian'] = df.julian.astype(int)
df['leap_year'] = df['date'].dt.is_leap_year
df = df[~df.index.astype(str).str.endswith('02-29')]  
n=0
for index,rows in df.iterrows():
    if df.leap_year[n] == True and df.new_julian[n] > 60:
        df.new_julian[n] = df.new_julian[n] - 1
    n=n+1
df
```
