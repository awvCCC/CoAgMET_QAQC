```python
akr02_tmax_trend = stats.linregress(akr02_df.tmax,akr02_df.maxt)
akr02_tmax_trend_start = 0*akr02_tmax_trend.slope+akr02_tmax_trend.intercept
akr02_tmax_trend_end = 100*akr02_tmax_trend.slope+akr02_tmax_trend.intercept
```

```python
akr02_df['date'] = akr02_df.date.astype(str)
akr02_df['max_temp_diff'] = akr02_df.max_temp_diff.astype(str)

akr02_tmaxdiff_fig = go.Figure(data=[go.Scatter()],
               layout=go.Layout(title=go.layout.Title(text='AKR02 - Akron 4 E Max Temp Diff'),
                               xaxis=go.layout.XAxis(title=go.layout.xaxis.Title(text='AKR02 Daily Maximum Temperature (°F)')),
                               yaxis=go.layout.YAxis(title=go.layout.yaxis.Title(text='Akron A 4 E Daily Maximum Temperature (°F)'))))
akr02_tmaxdiff_fig.add_trace(go.Scatter(x=akr02_df.tmax,y=akr02_df.maxt,
                                        text='Date: ' + akr02_df.date + "<br \>" 
                                        + "Difference: " + akr02_df.max_temp_diff,name='Max Temp',mode='markers'))
akr02_tmaxdiff_fig.add_trace(go.Scatter(x=[0,100],y=[akr02_tmax_trend_start,akr02_tmax_trend_end],
                                       mode='lines',name='Trendline',line=go.scatter.Line(color='blue')))
akr02_tmaxdiff_fig.add_trace(go.Scatter(x=[0,100],y=[0,100],mode='lines',name='1:1',
                                       line=go.scatter.Line(color='black')))
akr02_tmaxdiff_fig.show()
```
