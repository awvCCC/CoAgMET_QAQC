```python
def top_monthly_maxt(yearmonth):
    urllib.request.urlretrieve('http://data.rcc-acis.org/MultiStnData?state=CO&date=' + 
                               yearmonth + '&elems=mly_max_maxt&output=csv',filename=csv)

    data = pd.read_csv(csv)
    data.columns.values[0] = 'id'
    data.columns.values[1] = 'name'
    data.columns.values[2] = 'state'
    data.columns.values[3] = 'lat'
    data.columns.values[4] = 'long'
    data.columns.values[5] = 'elev'
    data.columns.values[6] = 'maxt'
    
    top_10_temps = data.nlargest(10,'maxt')
    
    return top_10_temps
```

```python
def lowest_monthly_mint(yearmonth):
    urllib.request.urlretrieve('http://data.rcc-acis.org/MultiStnData?state=CO&date=' + 
                               yearmonth + '&elems=mly_min_mint&output=csv',filename=csv)
    data = pd.read_csv(csv)
    data.columns.values[0] = 'id'
    data.columns.values[1] = 'name'
    data.columns.values[2] = 'state'
    data.columns.values[3] = 'lat'
    data.columns.values[4] = 'long'
    data.columns.values[5] = 'elev'
    data.columns.values[6] = 'mint'
    
    lowest_10_temps = data.nsmallest(10,'mint')
    
    return lowest_10_temps
```

```python
answer2 = lowest_monthly_mint('2020-06')
```

```python
def highest_daily_precip(yearmonth):
    urllib.request.urlretrieve('http://data.rcc-acis.org/MultiStnData?state=CO&date=' + 
                               yearmonth + '&elems=mly_max_pcpn&output=csv',filename=csv)
    data = pd.read_csv(csv)
    data.columns.values[0] = 'id'
    data.columns.values[1] = 'name'
    data.columns.values[2] = 'state'
    data.columns.values[3] = 'lat'
    data.columns.values[4] = 'long'
    data.columns.values[5] = 'elev'
    data.columns.values[6] = 'pcpn'
    
    data['pcpn'] = data['pcpn'].replace('T',0,regex=True)
    data['pcpn'] = data['pcpn'].replace('M',0,regex=True)
    data['pcpn'] = data.pcpn.astype(float)
    
    highest_10_rainfalls = data.nlargest(10,'pcpn')
    
    return highest_10_rainfalls
```

```python
answer3 = highest_daily_precip('2020-06')
```
