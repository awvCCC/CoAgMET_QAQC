```python
import os
import urllib
import pandas as pd
import numpy as np
```

```python
path = r'C:\Users\student\Documents\Alistair CoAgMET Projects\Pythonstuff\QC'
os.chdir(path)

csv = r'C:\Users\student\Documents\Alistair CoAgMET Projects\Pythonstuff\QC\csv.csv'
f = open(csv,'w+')
```

```python
urllib.request.urlretrieve(
    'http://coagmet.colostate.edu/cgi-bin/web_services.pl?type=daily&sids=akr02&sdate=1992-07-01&edate=2018-12-31&elems=tave,tmax,tmin',
    filename=csv)

df = pd.read_csv(csv)
df = df.reset_index()
df.columns = ['station','date','tave','tmax','tmin']
df['date'] = pd.to_datetime(df.date)
df.drop(df.tail(1).index, inplace=True)
df.tave = df.tave.astype(float)
df.tmin = df.tmin.astype(float)
df.tmax = df.tmax.astype(float)
print(df)
f=0
nulls = 0
for index,rows in df.iterrows():
    if(pd.isnull(df['tave'][f])):
        nulls = nulls+1
    f=f+1
print('There are ' + str(nulls) + ' missing average temperature observations in the dataset')
h=0
nulls=0
for index,rows in df.iterrows():
    if(pd.isnull(df['tmax'][h])):
        nulls=nulls+1
    h=h+1
print('There are ' + str(nulls) + ' missing maximum temperature observations in the dataset')
i=0
nulls=0
for index,rows in df.iterrows():
    if(pd.isnull(df['tmin'][i])):
        nulls = nulls + 1
    i=i+1
print('There are ' + str(nulls) + ' missing minimum temperature observations in the dataset')
length = len(df.index)

r = pd.date_range(start='1992-07-01',end='2018-12-31',freq='D')

nulls = r - length
print('There are ' + str(nulls) + 'entirely missing observations in the dataset')
df = df.set_index('date').reindex(r,copy=False).rename_axis('date')
print(df)
j=0
replaced = 0
for index,rows in df.iterrows():
    while(pd.isnull(df['tave'][j])) and(pd.notnull(df['tmax'][j])) and(pd.notnull(df['tmin'][j])):
        replaced = replaced + 1
        df.tave = (df.tmin + df.tmax) / 2
    j=j+1
print(str(replaced) + ' missing average temperature observations were replaced but not removed.')
length = len(df.index)
print('Currently the data set contains ' + str(length) + ' observations')
print()
df = df.dropna()
new_length = len(df.index)
null_observations = length - new_length
print(str(null_observations) + ' null observations were removed from the the data set' + ' now the dataset contains ' + str(new_length) + ' observations.')
print(df)
```
