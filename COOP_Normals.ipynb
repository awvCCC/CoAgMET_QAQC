```python
import os
import urllib
import pandas as pd
import numpy as np
```

```python
path = r'C:\Users\student\Documents\Alistair CoAgMET Projects\Pythonstuff\QC\minimums\COOP_minimums'
os.chdir(path)
xl_file = r"C:\Users\student\Documents\Alistair CoAgMET Projects\Pythonstuff\QC\colocation_metadata.xlsx"
csv = r'C:\Users\student\Documents\Alistair CoAgMET Projects\Pythonstuff\QC\normals\COOP_Normals\csv.csv'

xl = pd.ExcelFile(xl_file)

meta = xl.parse('metadata')
meta['sdate'] = meta.sdate.dt.date
meta['edate'] = meta.edate.dt.date
meta.drop(meta.columns[3],axis=1)


```

```python
n = 0
for index,rows in meta.iterrows():
    print('Grabbing COOP data for ' + str(meta.ID[n]))
    urllib.request.urlretrieve(
        'http://data.rcc-acis.org/StnData?sid=0' + str(meta.COOP_ID[n]) + '&sdate='+ str(meta.sdate[n]) + '&edate=' + str(meta.edate[n]) + '&elems=maxt,mint,avgt,pcpn&output=csv',
        filename=csv)
    print('Cleaning COOP data for ' + str(meta.ID[n]))
    df = pd.read_csv(csv)
    df = df.reset_index()
    df.columns.values[0] = 'date'
    df.columns.values[1] = 'maxt'
    df.columns.values[2] = 'mint'
    df.columns.values[3] = 'avgt'
    df.columns.values[4] = 'pcpn'
    df['date'] = pd.to_datetime(df.date)
    df['julian'] = df.date.dt.strftime('%j')
    df = df.replace('M', np.nan, regex=True)
    df = df.replace('T', 0.005, regex=True)
    df.maxt = df.maxt.astype(float)
    df.mint = df.mint.astype(float)
    df.avgt = df.avgt.astype(float)
    df.pcpn = df.pcpn.astype(float)
    print('Analyzing COOP data for ' + str(meta.ID[n]))
    normals = df.groupby(['julian']).min().reset_index()
    normals.to_csv(path + r'/' + str(meta.ID[n]) + '_COOPmin.csv')
    n = n+1
```
