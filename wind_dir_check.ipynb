```python
path = r'C:\Users\student\Documents\Alistair CoAgMET Projects\Pythonstuff\QC\windQC'

os.chdir(path)

today = datetime.now().strftime("%Y-%m-%d")
yesterday = datetime.strftime(datetime.now() - timedelta(1), "%Y-%m-%d")
         
txt = r'C:\Users\student\Documents\Alistair CoAgMET Projects\Pythonstuff\QC\windQC\daily_qc\wg_report_' + str(yesterday) + '.txt'
g = open(txt,'a')

csv = r'C:\Users\student\Documents\Alistair CoAgMET Projects\Pythonstuff\QC\windQC\csv.csv'
f = open(csv,'w+')
xl_file = r"C:\Users\student\Documents\Alistair CoAgMET Projects\Pythonstuff\QC\colocation_metadata.xlsx"
```

```python
xl = pd.ExcelFile(xl_file)

meta = xl.parse('metadata')
meta['sdate'] = meta.sdate.dt.date
meta['edate'] = meta.edate.dt.date
meat = meta.drop(meta.columns[3],axis=1)
```

```python
today = datetime.now().strftime("%Y-%m-%d")
yesterday = datetime.strftime(datetime.now() - timedelta(1), "%Y-%m-%d")
```

```python
n=0
for index,rows in meta.iterrows():
    try:
        print('Pulling wind direction data for ' + str(meta.ID[n]))
        urllib.request.urlretrieve(
        'http://coagmet.colostate.edu/cgi-bin/web_services.pl?type=five_minute&sids=' + str(meta.ID[n]) + '&sdate=' + str(yesterday) + '&edate=' + str(today) + '&elems=wind_vec,gustdir', 
        filename=csv)
        df = pd.read_csv(csv)
        f.close()
    except Exception as e:
        print('Something went wrong while pulling the wind direction data for ' + str(meta.ID[n]) + ': ' + e.args[0])
    
    try:
        print('Cleaning the wind direction data for ' + str(meta.ID[n]))
        df = df.reset_index()
        df.columns.values[0] = 'station'
        df.columns.values[1] = 'date'
        df.columns.values[2] = 'wind_dir'
        df.columns.values[3] = 'gust_dir'
        df.wind_dir.astype(float)
        df.gust_dir.astype(float)
    except Exception as e:
        print('Something went wrong while cleaning the wind direction data for ' + str(meta.ID[n]) + ': ' + e.args[0])

    except Exception as e:
        print('Something went wrong while cerating a text file for ' + str(meta.ID[n]) + ': ' + e.args[0])

    try:
        print('Checking wind direction data from ' + str(meta.ID[n]) + ' for negative values...')
        d=0
        for index,rows in df.iterrows():
            if df.wind_dir.any() < 0:
                string = (str(df.station[n]) + ' reported negative wind direction at ' + str(df.date[n]))
                g.write(string)
            else:
                continue
            d=d+1
    except Exception as e:
        print('Something went wrong while cheking the wind data from ' + str(meta.ID[n]) + ': ' + e.args[0])
    n=n+1
print('Wind data succesfully checked')
g.close()
```
