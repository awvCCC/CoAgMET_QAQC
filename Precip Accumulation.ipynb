```python
import os
from datetime import datetime,timedelta
import urllib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
today = datetime.now().strftime("%Y-%m-%d")
path = r'C:\Users\student\Documents\Alistair CoAgMET Projects\Pythonstuff\QC\precip_QC\2019WY'

os.chdir(path)
os.mkdir(path + r'\accumulated_precip_' + str(today))

figpath = r'C:\Users\student\Documents\Alistair CoAgMET Projects\Pythonstuff\QC\precip_QC\2019WY\accumulated_precip_' + str(today)

csv = r'C:\Users\student\Documents\Alistair CoAgMET Projects\Pythonstuff\QC\precip_QC\csv.csv'
f = open(csv,'w+')
wy_sdate = '2018-10-01'

sns.set_style('dark')
sns.set(rc={'figure.figuresize' : (11.7,8.27)})
```

```python
arkCollegiateValley = 'bnv01,sld01'
arkEastValley = 'hly01,hly02,lam01,lam03,lam04,mcl01'
arkMesas = 'cnn01,hne01,pnr01,wcf01'
arkNorthofValley = 'scm01'
arkPlateauSouthofValley = 'wls01'
arkWestValleyBottom = 'avn01,fwl01,ljt01,rfd01'

coFourCorners = 'ctz01,dvc01,mnc01,twc01,yjk01,yuc01'
coHeadwaters = 'gby01,krm01,wfd01'
coLowerGunnison = 'dlt01,ekt01,hot01,mtr01,nwd01,oth01'
coLowerValley = 'bdk01,gyp01,cbl01,frt02,orm01,cbn01,slt01'
coSanJuan = 'drg01,ign01,kln01,pgs01'
coUpperGunnsion = 'gun01'
coYampaWhite = 'clk01,hyd01,mkr01'

ksNorthPlains = 'akr02,hxt01,hyk02,ilf01,pai01,stg01,wry01,yum02'
ksSouthPlains = 'brl02,brl03,idl01,krk01,sbt01,stn01'

pltFrontRangefoothills = 'cpk01'
pltLowerPlains = 'brg01,hxt01,ksy01,ksy02'
pltNorthFrontRange = 'alt01,ftc01,ftc03,gly04,lcn01'
pltNorthPark = 'cow01,heb01,lar01'
pltSouthFrontRagne = 'eac01,flt01,lsl01,pkh01,pkr01'
pltSouthPark = 'bnv01,jfn01,sld01'

rioSanLuisValley = 'bla01,ctr01,ctr02,ljr01,san01'
```

```python
print('Creating a precip accumulation plot for the Collegiate Valley of the Arkansas Drainage...')
try:
    print('Pulling precip data from the Collegiate Valley of the Arkansas Drainage...')
    urllib.request.urlretrieve('http://coagmet.colostate.edu/cgi-bin/web_services.pl?type=daily&sids=' + str(arkCollegiateValley)
                              + '&sdate=' + str(wy_sdate) + '&edate=' + str(today) + '&elems=pp',
                             filename=csv)
    df = pd.read_csv(csv)
    f.close()
except Exception as e:
    print('Something went wrong while pulling the precip data from the Collegiate Valley of the Arkansas Drainage: ' + e.args[0])
try:
    print('Cleaning the precip data from the Collegiate Valley of the Arkansas Drainage...')
    df = df.reset_index()
    df.columns.values[0] = 'station'
    df.columns.values[1] = 'date'
    df.columns.values[2] = 'precip'
    df.precip.astype(float)
    df.precip = df.precip / 25.4
except Exception as e:
    print('Something went wrong while cleaning the precip data from the Collegiate Valley of the Arkansas Drainage: ' + e.args[0])
try:
    print('Creating an accumulated precip column using the precip data from the Collegiate Valley of the Arkansas Drainage...')
    df['accumulated_precip'] = df.groupby('station')['precip'].cumsum()
except Exception as e:
    print('Something went wrong while creating an accumulated precip column using the precip data from the Collegiate Valley of the Arkansas Drainage: '
         + e.args[0])
try:
    print('Creating an accumulation plot using Collegiate Valley precip data...')
    plot = sns.lineplot(x=df.date,y=df.accumulated_precip,hue=df.station,data=df,palette='colorblind')
    plot.set(xlabel='Date',ylabel='Accumulated Precip for the 2019 WY')
    title = str('Arkansas Drainage - Collegiate Valley')
    plot.xaxis.set_major_locator(plt.MaxNLocator(6))
    plt.savefig(str(figpath) + r'\Arkansas_Collegiate_Valley_' + str(today) + '.png')
    plt.close()
except Exception as e:
    print('Something went wrong while creating the accumulation plot form Collegiate Valley Precip Data: ' + e.args[0])
print('Creating a precip accumulation plot for the East Valley Bottom of the Arkansas Drainage')
try:
    print('Pulling precip data from the East Valley Bottom of the Arkansas Drainage...')
    f.open()
    urllib.request.urlretrieve('http://coagmet.colostate.edu/cgi-bin/web_services.pl?type=daily&sids=' + str(arkEastValley)
                              + '&sdate=' + str(wy_sdate) + '&edate=' + str(today) + '&elems=pp',
                             filename=csv)
    df = pd.read_csv(csv)
    f.close()
except Exception as e:
    print('Something went wrong while pulling precip data from the East Valley Bottom of the Arkansas Drainage: ' + e.args[0])
try:
    print('Cleaning the precip data from the East Valley Bottom of the Arkansas Drainage...')
    df = df.reset_index()
    df.columns.values[0] = 'station'
    df.columns.values[1] = 'date'
    df.columns.values[2] = 'precip'
    df.precip.astype(float)
    df.precip = df.precip / 25.4
except Exception as e:
    print('Something went wrong while cleaning the precip data from the East Valley Bottom of the Arkansas Drainage: ' + e.args[0])
try:
    print('Creating an accumulated precip column using the precip data from the East Valley Bottom of the Arkansas Drainage: ' + e.args[0])
    df['accumulated_precip'] = df['precip'].cumsum()
except Exception as e:
    print('Something went wrong while creating a accumulated precip column using data from the East Valley Bottom of the Arkansas Drainage: ' + e.args[0])
try:
    print('Creating an accumulation plot using the East Valley Bottom precip data...')
    plot = sns.lineplot(x=df.date,y=df.accumulate_precip,hue=df.station,data=df,palette='colorblind')
    plot.set(xlabel='Date',ylabel='Accumulated Precip for the 2019 WY')
    title = str('Arkansas Drainage - East Valley Bottom')
    plot.xaxis.set_major_locator(plt.MaxNLocator(6))
    plt.legend()
    plt.savefig(str(figpath) + r'\Arkansas_East_Valley_Bottom_' + str(today) + '.png')
    plt.clf()
except Exception as e:
    print('Something went wrong while creating an accumulation plot from East Valley Bottom precip data: ' + e.args[0])
print('Creating a precip accumulation plot for Mesas of the Arkansas Drainage')
try:
    print('Pulling Precip data from the Mesas of the Arkansas Drainage...')
    f.open()
    urllib.request.urlretrieve('http://coagmet.colostate.edu/cgi-bin/web_services.pl?type=daily&sids=' + str(arkMesas)
                              + '&sdate=' + str(wy_sdate) + '&edate=' + str(today) + '&elems=pp',
                             filename=csv)
    df = pd.read_csv(csv)
    f.close()
except Exception as e:
    print('Something went wrong while pulling precip data from the Mesas of the Arkansas Drainage: ' + e.args[0])
try:
    print('Cleaning precip data from the Mesas of the Arkansas Drainage...')
    df = df.reset_index()
    df.columns.values[0] = 'station'
    df.columns.values[1] = 'date'
    df.columns.values[2] = 'precip'
    df.precip.astype(float)
    df.precip = df.precip / 25.4
except Exception as e:
    print('Something went wrong while cleaning precip data from the Mesas of the Arkansas Drainage...')
try:
    print('Creating an accumulated precip column using precip data from the Mesas of the Arkansas Drainage...')
    df['accumulated_precip'] = df['precip'].cumsum()
except Exception as e:
    print('Something went wrong while creating an accumulated precip column using data from the Mesas of the Arkansas Drainage: ' + e.args[0])
try:
    print('Creating an accumulatin plot using the Mesas precip data...')
    plot = sns.lineplot(x=df.date,y=df.accumulate_precip,hue=df.station,data=df,palette='colorblind')
    plot.set(xlabel='Date',ylabel='Accumulated Precip for the 2019 WY')
    title = str('Arkansas Drainage - Mesas')
    plot.xaxis.set_major_locator(plt.MaxNLocator(6))
    plt.legend()
    plt.savefig(str(figpath) + r'\Arkansas_Mesas_' + str(today) + '.png')
    plt.clf()
except Exception as e:
    print('Something went wrong while creating an accumulation plot from the Mesas precip data: ' + e.args[0])
print('Creating an accumulation plot using precip data from the North of Valley region of the Arkansas Drainage...')
try:
    print('Pulling precip data from the North of Valley region of the Arkansas Drainage...')
    f.open()
    urllib.request.urlretrieve('http://coagmet.colostate.edu/cgi-bin/web_services.pl?type=daily&sids=' + str(arkNorthofValley)
                              + '&sdate=' + str(wy_sdate) + '&edate=' + str(today) + '&elems=pp',
                             filename=csv)
    df = pd.read_csv(csv)
    f.close()
except Exception as e:
    print('Something went wrong while pulling precip data from the North of Valley region of the Arkansas Drainage: ' + e.args[0])
try:
    print('Cleaning precip data from the North of Valley region of the Arkansas Drainage...')
    df = df.reset_index()
    df.columns.values[0] = 'station'
    df.columns.values[1] = 'date'
    df.columns.values[2] = 'precip'
    df.precip.astype(float)
    df.precip = df.precip / 25.4
except Exception as e:
    print('Something went wrong while cleaning the precip data from the North of Valley region of the Arkansas Drainage: ' + e.args[0])
try:
    print('Creating an accumulated precip column using the precip data from the North of Valley region of the Arknasas Drainage...')
    df['accumulated_precip'] = df['precip'].cumsum()
except Exception as e:
    print('Something went wrong while creating an accumulated precip column using data from the North of Valley region of Arkansas Drainage: ' + e.args[0])
try:
    print('Creating an accumulation plot from the North of Valley precip data...')
    plot = sns.lineplot(x=df.date,y=df.accumulate_precip,hue=df.station,data=df,palette='colorblind')
    plot.set(xlabel='Date',ylabel='Accumulated Precip for the 2019 WY')
    title = str('Arkansas Drainage - North of Valley')
    plot.xaxis.set_major_locator(plt.MaxNLocator(6))
    plt.legend()
    plt.savefig(str(figpath) + r'\Arkansas_North_of_Valley_' + str(today) + '.png')
    plt.clf()
except Exception as e:
    print('Something went wrong while creating an accumulation plot fromt the North of Valley precip data: ' + e.args[0])
print('Creating an accumulation plot using precip data from the Plateau South of Valley region of the Arkansas Drainage...')
try:
    print('Pulling precip data from the Pleatuea South of Valley Region of the Arkansas Drainage...')
    f.open()
    urllib.request.urlretrieve('http://coagmet.colostate.edu/cgi-bin/web_services.pl?type=daily&sids=' + str(arkPlateauSouthofValley)
                              + '&sdate=' + str(wy_sdate) + '&edate=' + str(today) + '&elems=pp',
                             filename=csv)
    df = pd.read_csv(csv)
    f.close()
except Exception as e:
    print('Something went wrong while pulling data from the Plateau South of Vallyey region of the Arkansas Drainage: ' + e.args[0])
try:
    print('Cleaning precip data from the Plateau South of Valley region of the Arkansas Drainage...')
    df = df.reset_index()
    df.columns.values[0] = 'station'
    df.columns.values[1] = 'date'
    df.columns.values[2] = 'precip'
    df.precip.astype(float)
    df.precip = df.precip / 25.4
except Exception as e:
    print('Something went wrong while cleaning the precip data from the Plateau South of Valley region of the Arkansas Drainage'
         + e.args[0])
    
    
```
